```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
# Get dataset
df_sal = pd.read_csv('Salary_Data.csv')
df_sal.head(6)
```

```python
# Describe data
            df_sal.describe()
```

```python
df_sal.info()
```

```python
sns.displot(df_sal['Salary'])
plt.title('Salary Distribution Plot')
plt.show()
```

```python
df_sal.head()
```

```python
plt.scatter(df_sal['YearsExperience'], df_sal['Salary'], color='green')
plt.title('Salary vs Exp')
plt.xlabel('Year of Exp')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
X = df_sal.iloc[:, :1]  # independent
y = df_sal.iloc[:, 1:]  # dependent
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test) # predicted value of y_test
y_pred_train = regressor.predict(X_train)  # predicted value of y_train
```

```python
plt.scatter(X_train, y_train, color='lightgreen')
plt.plot(X_train, y_pred_train, color='g')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
```

```python
plt.scatter(X_test, y_test, color='lightgreen')
plt.plot(X_train, y_pred_train, color='g')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python

print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
